```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
```

```python
# Dummy dataset
data = {
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Clara", "David", "Ella", "Frank", "Grace", "Henry", "Ivy", "Jack"],
    "Gender": ["Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [55000, 60000, 58000, 65000, 56000, 62000, 57000, 67000, 54000, 63000],
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate average salary by gender
average_salary_by_gender = df.groupby("Gender")["Salary"].mean()

# Calculate pay gap percentage
pay_gap_percentage = (
    (average_salary_by_gender["Male"] - average_salary_by_gender["Female"])
    / average_salary_by_gender["Male"]
) * 100

# Display insights
print("Average Salary by Gender:")
print(average_salary_by_gender)
print(f"\nPay Gap Percentage: {pay_gap_percentage:.2f}%\n")

# Visualization
plt.figure(figsize=(8, 5))
average_salary_by_gender.plot(kind="bar", color=["pink", "blue"], edgecolor="black")
plt.title("Average Salary by Gender", fontsize=14)
plt.ylabel("Average Salary", fontsize=12)
plt.xlabel("Gender", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
